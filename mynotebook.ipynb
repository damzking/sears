```python
# Create a complete date range from Jan 6, 2004, to Nov 30, 2023
date_range = pd.date_range(start='2004-01-06', end='2023-11-30')
```

```python
# Filter the data based on the date range for each currency
df_filtered = df[
    (df['Date_GBP'].isin(date_range)) &
    (df['Date_usd'].isin(date_range)) &
    (df['Date_EUR'].isin(date_range))
]
```

```python
df_filtered
```

```python
# List of weekend days to check for
weekend_days = ['Saturday', 'Sunday']

# Columns to check
day_columns = ['Day Name_fOR_USD', 'Day_ Name_EUR', 'Day_Name_GBP']

# Use a for loop to iterate through each column and drop rows that contain 'Saturday' or 'Sunday'
for column in day_columns:
    df_filtered = df_filtered[~df_filtered[column].isin(weekend_days)]  # ~ negates the condition to drop rows with weekends

# Display the filtered DataFrame
print(df_filtered)
```
